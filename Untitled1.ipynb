```python
import pandas as pd 
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objs as go
```

```python
data = pd.read_csv('final_ff_1.csv', on_bad_lines='skip', sep=';')
data.head()
```

```python
data.day_of_week.value_counts()
```

```python
data.info()
```

```python
data.date_.value_counts()
```

```python
data = data[~(data.user_hour.isna())]
```

```python
for i in range(data.shape[0]): 
    line = data.iloc[i, 14] 
    line = line[:len(line) - 2] 
    data.iloc[i, 14] = int(line)
data.head()
```

```python
for i in range(data.shape[0]):
    if data.iloc[i, 1] == '2022-10-16': 
        data.iloc[i, 13] = 7
    elif data.iloc[i, 1] == '2022-10-17': 
        data.iloc[i, 13] = 1
    elif data.iloc[i, 1] == '2022-10-18':
        data.iloc[i, 13] = 2
    elif data.iloc[i, 1] == '2022-10-20': 
        data.iloc[i, 13] = 4
    elif data.iloc[i, 1] == '2022-10-21': 
        data.iloc[i, 13] = 5
    else: 
        data.iloc[i, 13] = 6
```

```python
data.day_of_week.value_counts()
```

```python
for i in range(data.shape[0]):
    if data.iloc[i, 15] > 0: 
        data.iloc[i, 15] = 1 
data.button_tap_flg.value_counts()
```

```python
data = data[(data.age <= 70) & (data.age >= 14)]
```

```python
data['index'] = data['button_tap_flg'] + data['like_tap_flg'] + data['share_tap_flg'] + data['favorite_tap_flg']
```

```python
hours_act = data.groupby('user_hour', as_index=False).agg({'index': 'sum'}).sort_values('user_hour') 
hours_act
```

```python
corr1 = np.corrcoef(list(hours_act.user_hour), list(hours_act.index))
corr1
```

```python
hours_count = data.user_hour.value_counts()
hours_count
```

```python
hours_act['otn_index'] = 0 * hours_act.shape[0] 
```

```python
for i in range(hours_act.shape[0]): 
    hours_act.iloc[i, 2] += round(hours_act.iloc[i, 1] / hours_count[i], 3)
hours_act
```

```python
#график по абсолютному индексу
plt.scatter(x=hours_act['user_hour'], y=hours_act['index'], color='green') 
plt.xlabel('Час') 
plt.ylabel('Абсолютный индекс активности') 
plt.title('Зависимость абсолютного индекса активности от времени суток')
plt.show()
```

```python
#график по относительному индексу активности
plt.scatter(x=hours_act['user_hour'], y=hours_act['otn_index'], color='red')
plt.yticks([0, 0.04, 0.08, 0.12, 0.16, 0.2])
plt.title('Зависимость относительного индекса активности от времени суток')
plt.xlabel('Час') 
plt.ylabel('Относительный индекс активности')
plt.show()
```

```python
print('Корреляция между абсолютным и относительным индексами активности:', round(np.corrcoef(list(hours_act.index), list(hours_act.otn_index))[0][1], 2))
```

```python
#из графика мы можем сделать вывод о том, что в позднее время(вечер) ноблюдается такая же активность, как я днём
#поэтому гипотеза опровергнута, так же можно заметить, что в 5 часов утра наблюдается падение активности
```

```python
gender_act = data.groupby('gender', as_index=False).agg({'index': 'sum'})
gender_counts = data.gender.value_counts() 
gender_act['otn_index'] = [0] * gender_act.shape[0]
for i in range(gender_act.shape[0]): 
    gender_act.iloc[i, 2] += gender_act.iloc[i, 1] / gender_counts[gender_act.iloc[i, 0]]
gender_act
```

```python
sns.histplot(data.age, bins=len(data.age.unique())).set(title='Гистограмма по количетсу возрастов', xlabel='Возраст', ylabel='Количество')
```

```python
age_act = data.groupby('age', as_index=False).agg({'index': 'sum'}).sort_values('age') 
age_counts = data.age.value_counts() 
age_act['otn_index'] = [0] * age_act.shape[0] 
for i in range(age_act.shape[0]): 
    age_act.iloc[i, 2] += age_act.iloc[i, 1] / age_counts[age_act.iloc[i, 0]]
age_act
```

```python
#график изменения активности по возрасту 
plt.scatter(x=age_act['age'], y=age_act['index'], color='red')
plt.title('Зависимость между возрастом и абсолютной активностью')
plt.xlabel('Возраст')
plt.ylabel('Абсолютная активность')
```

```python
plt.scatter(x=age_act['age'], y=age_act['otn_index'], color='red')
plt.yticks([0, 0.04, 0.08, 0.12, 0.16, 0.2])
plt.title('Зависимость между относительной активностью и возрастом')
plt.xlabel('Возраст') 
plt.ylabel('Относительная активность')
```

```python
#корреляция между возрастом и относительной активностью 
print(round(np.corrcoef(list(age_act.age), list(age_act.otn_index))[0][1], 2))
```

```python
#из данного графика можно заметить, что в возрасте +-35 лет пик относительной активности, в то время как у молодых пользователей
#и полжилых пользователей относительная активность падает, вероятнее всего это связао с тем, что люди в молодом и пожилом 
#возрасте не сильно важно содержание сториса, поэтому они не проявляют активности, а смотрят из-за интереса
#а люди в возрасте +-35 вероятнее всего работают, поэтому содержание сторис имеет для них важность, поэтому они проявляют активность
```
